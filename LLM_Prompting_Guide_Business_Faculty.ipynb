````python
import anthropic

client = anthropic.Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

def run_prompt(prompt, system=None, max_tokens=1024):
    """Helper function to run a prompt and display the response."""
    kwargs = {
        "model": "claude-opus-4-5",
        "max_tokens": max_tokens,
        "messages": [{"role": "user", "content": prompt}]
    }
    if system:
        kwargs["system"] = system

    message = client.messages.create(**kwargs)
    response = message.content[0].text
    print(response)
    return response

print("✓ Claude client ready! Use run_prompt(your_prompt) throughout this notebook.")
````

## 1.1 What is Prompting?

**Prompting** is the art and science of communicating with Large Language Models (LLMs) to get the best possible outputs. Think of it as giving clear instructions to a highly capable assistant.

### Why Different Models?

| Model | Company | Best For |
|---|---|---|
| **Claude** | Anthropic | Long documents, detailed analysis, coding, ethical reasoning |
| **Gemini** | Google | Multimodal (text + images), integrated with Google Workspace |
| **ChatGPT / GPT-4** | OpenAI | Versatile, widely integrated, strong creative capabilities |

### Key Principle:
> **Garbage In = Garbage Out** | **Quality In = Quality Out**  
> The better your prompt, the better the response!

# Part 2: General Prompting Best Practices
---

## 2.1 The CLEAR Framework

| Letter | Stands For | Meaning |
|---|---|---|
| **C** | Context | Provide background information |
| **L** | Length | Specify desired output length |
| **E** | Examples | Give examples when possible |
| **A** | Audience | Define who the output is for |
| **R** | Role | Assign a role to the AI |

### Basic Prompt Structure:

```
[Role] + [Context] + [Task] + [Format] + [Constraints]
```


## 2.2 Poor vs. Good Prompts — Live Comparison

Run each cell below and compare the outputs side by side.


````python
# ❌ POOR PROMPT — vague, no context
poor_prompt = "Write about finance"

print("=" * 60)
print("POOR PROMPT OUTPUT:")
print("=" * 60)
run_prompt(poor_prompt, max_tokens=256)
````

## 3.1 Financial Analysis

**Why This Works:**
- Defines role (financial analyst)
- Specifies analysis dimensions
- Requests structured output

````python
financial_analysis_prompt = """You are a senior financial analyst. Below are sample excerpts from a hypothetical
tech company's latest annual filing:

- Revenue grew from $45B to $51B (+13% YoY)
- Operating income: $12B (margin: 23.5%)
- Free cash flow: $9B
- Long-term debt: $25B; Cash & equivalents: $18B
- Key risks cited: supply chain disruption, regulatory scrutiny in EU, FX headwinds

Please analyze the following aspects:
1. Revenue trends and growth patterns
2. Operating margin analysis
3. Key risks mentioned
4. Capital allocation strategy

Present your analysis in a structured format with clear sections.
Highlight 3 key insights most relevant for potential investors."""

run_prompt(financial_analysis_prompt)
````

### Task: Creating Financial Models (DCF)

**Claude's Strength**: Shows detailed calculations with step-by-step explanations — ideal for teaching.


````python
dcf_prompt = """I need to create a 5-year DCF model for a mid-sized retail company.

Company Details:
- Current Revenue: $50M
- Expected growth: 8% annually years 1-3, then 5% years 4-5
- Operating margin: 12%
- Tax rate: 25%
- WACC: 9%

Please help me:
1. Calculate projected free cash flows for each year
2. Calculate terminal value using perpetuity growth of 3%
3. Discount all cash flows to present value
4. Provide the enterprise value

Show your work step by step, and explain any assumptions made."""

run_prompt(dcf_prompt)
````

# Part 8: Common Pitfalls and How to Avoid Them
---

## The 8 Most Common Prompting Mistakes

Each cell below demonstrates a pitfall — run both the bad and good version to see the difference.

---

### ❌ Pitfall #1: Vague Prompts


````python
# ❌ Bad — too vague
bad = "Tell me about marketing"

# ✅ Good — specific, with audience, format, and length
good = """Explain the 4 Ps of marketing (Product, Price, Place, Promotion)
with modern examples from tech companies. Target audience is undergraduate
business students with no prior marketing knowledge.
Aim for 500 words with 2 examples per P."""

print("=" * 60)
print("✅ GOOD PROMPT OUTPUT:")
print("=" * 60)
run_prompt(good)
````

### ❌ Pitfall #2: No Context or Background


````python
# ✅ Good — context-rich investment query
investment_prompt = """I'm considering investing in Company X stock.

Context:
- Investment horizon: 5 years
- Risk tolerance: Moderate
- Portfolio: Currently 60% stocks, 40% bonds
- Goal: Retirement savings

Company X details:
- P/E ratio: 25
- Dividend yield: 2%
- 5-year growth: 15% annually
- Sector: Technology
- Market cap: $50B

Analyze whether this fits my investment profile and goals."""

run_prompt(investment_prompt)
````

### ❌ Pitfall #5: One-Shot Prompting — The Power of Iteration

Run the cells below **in sequence** to see how iteration improves outputs.


````python
# Step 1: Start with an outline
prompt_1 = """Create an outline for a presentation on cryptocurrency
for traditional finance professionals."""

print("STEP 1 — Outline:")
outline = run_prompt(prompt_1)
````

````python
# Step 2: Expand a specific section
prompt_2 = f"""Here is my presentation outline:
{outline}

Now expand the section on blockchain technology.
Make it more accessible to a non-technical audience of traditional bankers.
Use an analogy they would relate to."""

print("STEP 2 — Expanded blockchain section:")
expanded = run_prompt(prompt_2)
````

````python
# Step 3: Add real-world examples
prompt_3 = f"""Based on this expanded section:
{expanded}

Add 3 case studies of real-world blockchain applications in traditional banking.
Format each as: Institution | Use Case | Outcome (2-3 sentences each)."""

print("STEP 3 — With case studies:")
run_prompt(prompt_3)
````

### ❌ Pitfall #8: Expecting Real-Time Data

LLMs have a knowledge cutoff. Instead of asking for current stock prices, ask for frameworks and checklists.


````python
# ✅ Workaround — framework-based approach instead of real-time data
framework_prompt = """Based on historical patterns and company fundamentals,
what factors should I monitor when evaluating a large-cap technology stock?

Create a checklist of:
- Financial metrics to track
- Industry trends to watch
- Competitive dynamics to consider
- Macroeconomic indicators relevant to tech companies"""

run_prompt(framework_prompt)
````

## 9.1 Eight Power Techniques


````python
# 🎯 Tip 1: "Acts As" Framing
acts_as_prompt = """Act as a skeptical CFO reviewing a startup's pitch deck.

The startup claims:
- $5M revenue in Year 1 with zero marketing spend
- 80% gross margins in a competitive SaaS market
- Seeking $10M at a $100M valuation

What are your top 5 questions and concerns?"""

print("🎯 Tip 1: 'Acts As' Framing")
run_prompt(acts_as_prompt)
````

````python
# 🎯 Tip 2: Request Alternative Perspectives
perspectives_prompt = """A retail company is deciding whether to open a new flagship store
in a premium urban location at $2M annual rent.

Analyze this decision from three perspectives:
1. As a risk-averse CFO
2. As a growth-focused CEO
3. As a long-term shareholder

Show how each would view it differently and what they would recommend."""

print("🎯 Tip 2: Multiple Perspectives")
run_prompt(perspectives_prompt)
````

````python
# 🎯 Tip 3: Use Constraints Creatively
constraints_prompt = """Explain portfolio diversification using ONLY food analogies.
Make it memorable enough that an MBA student would never forget it.
Maximum 200 words."""

print("🎯 Tip 3: Creative Constraints")
run_prompt(constraints_prompt)
````

````python
# 🎯 Tip 5: Request Self-Critique
draft_analysis = """Our company should enter the Asian market next year because Asia has
a large population and growing middle class. Our products are high quality
and competitive pricing will help us win market share."""

critique_prompt = f"""Here is a business analysis draft:
\"\"\"
{draft_analysis}
\"\"\"

Critique this analysis:
1. What assumptions might be flawed?
2. What alternatives weren't considered?
3. What data would strengthen it?
4. What would a skeptic challenge?"""

print("🎯 Tip 5: Self-Critique")
run_prompt(critique_prompt)
````

````python
# 🎯 Tip 7: Chain of Thought
cot_prompt = """A company has the following situation:
- Current net income: $8M
- Shares outstanding: 10M
- They want to do a 15% stock buyback
- Price per share: $40

Solve this step by step:
1. First, calculate the cost of the buyback
2. Then, determine the new share count
3. Next, calculate the new EPS
4. Finally, assess whether this is value-enhancing for shareholders

Show your reasoning at each step."""

print("🎯 Tip 7: Chain of Thought")
run_prompt(cot_prompt)
````
