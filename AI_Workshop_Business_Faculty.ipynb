# Part 2: The Evolution of AI

## 2.1 The Four Eras of AI

### 📋 **Era 1: Symbolic AI (1950s-1980s)**
**Approach**: Rule-based systems with explicit programming

**Example in Business**:
```
IF customer_purchase > $1000 AND loyalty_years > 5
   THEN offer_discount = 20%
ELSE IF customer_purchase > $500
   THEN offer_discount = 10%
```

**Limitations**:
- Required manual rule creation
- Couldn't handle ambiguity
- Difficult to scale

---

### 📊 **Era 2: Machine Learning (1990s-2000s)**
**Approach**: Systems learn patterns from data

**Example in Business**:
- Credit scoring models
- Customer churn prediction
- Fraud detection

**Key Innovation**: The machine learns rules from examples instead of being explicitly programmed

---

### 🧠 **Era 3: Deep Learning (2010s)**
**Approach**: Multi-layered neural networks that learn hierarchical representations

**Example in Business**:
- Image recognition for quality control
- Speech recognition for customer service
- Recommendation systems

**Key Innovation**: Can automatically learn complex features without manual feature engineering

---

### 🤖 **Era 4: Large Language Models (2020s)**
**Approach**: Massive neural networks trained on enormous text datasets

**Examples**: GPT-5, Claude, Gemini, LLaMA

**Business Applications**:
- Automated report writing
- Customer service chatbots
- Market analysis
- Contract review
- Financial forecasting

**Key Innovation**: General-purpose intelligence that can handle diverse tasks with minimal training


## 2.2 Visualizing the Evolution

Let's create a simple visualization of AI capability growth:

````python
import matplotlib.pyplot as plt
import numpy as np

# Years and AI eras
years = np.array([1960, 1970, 1980, 1990, 2000, 2010, 2015, 2020, 2025])
capability = np.array([10, 15, 20, 30, 45, 65, 75, 85, 95])

plt.figure(figsize=(12, 6))
plt.plot(years, capability, 'b', linewidth=2, marker='o', markersize=8)

# Mark different eras
plt.axvspan(1950, 1985, alpha=0.2, color='red', label='Symbolic AI')
plt.axvspan(1985, 2010, alpha=0.2, color='green', label='Machine Learning')
plt.axvspan(2010, 2018, alpha=0.2, color='blue', label='Deep Learning')
plt.axvspan(2018, 2026, alpha=0.2, color='purple', label='LLMs')

plt.xlabel('Year', fontsize=12)
plt.ylabel('AI Capability (Relative)', fontsize=12)
plt.title('Evolution of AI: From Rules to Intelligence', fontsize=14, fontweight='bold')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

# Part 3: Hands-On Neural Network Example
---

Now let's get hands-on! We'll build a simple neural network to understand how modern AI learns.

## 3.1 The MNIST Dataset: A Classic Learning Problem

**MNIST** = Modified National Institute of Standards and Technology database

- 70,000 handwritten digit images (0-9)
- Each image is 28×28 pixels (784 numbers)
- Classic "Hello World" of machine learning

### Business Analogy:
Think of this like teaching a system to read handwritten numbers on checks or forms - a real business problem!


## 3.2 Setting Up Our Environment

````python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
torch.manual_seed(42)

print("✓ Libraries imported successfully!")
print(f"✓ PyTorch version: {torch.__version__}")
````

## 3.3 Loading and Visualizing the Data

````python
# Define data transformations
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # MNIST mean and std
])

# Download and load training data
train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

# Download and load test data
test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

print(f"✓ Training samples: {len(train_dataset):,}")
print(f"✓ Test samples: {len(test_dataset):,}")
````

````python
# Visualize some examples
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
fig.suptitle('Sample Handwritten Digits from MNIST', fontsize=14, fontweight='bold')

for i, ax in enumerate(axes.flat):
    image, label = train_dataset[i]
    ax.imshow(image.squeeze(), cmap='gray')
    ax.set_title(f'Label: {label}', fontsize=12)
    ax.axis('off')

plt.tight_layout()
plt.show()

print("\nEach image is a 28×28 pixel grid (784 total pixels)")
print("Each pixel has a value from 0 (white) to 1 (black)")
````

## 3.4 Preparing Data for Training

````python
# Create data loaders (these help feed data to our model in batches)
batch_size = 64

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False
)

print(f"✓ Created data loaders with batch size: {batch_size}")
print(f"  Number of training batches: {len(train_loader)}")
print(f"  Number of test batches: {len(test_loader)}")
````

## 3.5 Defining Our Neural Network

We'll create a **Multi-Layer Perceptron (MLP)** - a simple neural network with:
- **Input layer**: 784 neurons (one per pixel)
- **Hidden layer**: 128 neurons (learns patterns)
- **Output layer**: 10 neurons (one per digit 0-9)

### Business Analogy:
Think of this like a decision-making pipeline:
1. **Input**: Raw data (pixels)
2. **Processing**: Extract patterns (hidden layer)
3. **Decision**: Classification (output)


````python
# Define the neural network architecture
class SimpleMLP(nn.Module):
    def __init__(self):
        super(SimpleMLP, self).__init__()
        # Input: 784 pixels (28x28)
        # Hidden: 128 neurons
        # Output: 10 classes (digits 0-9)

        self.fc1 = nn.Linear(28 * 28, 128)  # First layer
        self.relu = nn.ReLU()               # Activation function
        self.fc2 = nn.Linear(128, 10)       # Output layer

    def forward(self, x):
        # Flatten the image
        x = x.view(-1, 28 * 28)
        # Pass through layers
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Create the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleMLP().to(device)

print("✓ Neural Network Architecture:")
print(model)
print(f"\n✓ Model is using: {device}")

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"\n✓ Total trainable parameters: {total_params:,}")
print("  (These are the 'weights' the model will learn)")
````

## 3.6 Setting Up Training

We need to define:
- **Loss function**: Measures how wrong our predictions are
- **Optimizer**: Adjusts the model to reduce errors
- **Learning rate**: How big of adjustments to make

````python
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training parameters
num_epochs = 5  # How many times to go through all the data

print("✓ Training Configuration:")
print(f"  Loss Function: Cross Entropy")
print(f"  Optimizer: Adam")
print(f"  Learning Rate: 0.001")
print(f"  Number of Epochs: {num_epochs}")
print(f"  Batch Size: {batch_size}")
````

## 3.7 Training the Model

Now we'll train our neural network. This is where the "learning" happens!

**What's happening:**
1. Show the model examples
2. Model makes predictions
3. Calculate error
4. Adjust model weights to reduce error
5. Repeat!


````python
# Training loop
train_losses = []
train_accuracies = []

print("Starting training...\n")

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for batch_idx, (data, target) in enumerate(train_loader):
        # Move data to device
        data, target = data.to(device), target.to(device)

        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        output = model(data)
        loss = criterion(output, target)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        # Statistics
        running_loss += loss.item()
        _, predicted = torch.max(output.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

    # Calculate epoch statistics
    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total
    train_losses.append(epoch_loss)
    train_accuracies.append(epoch_acc)

    print(f"Epoch {epoch+1}/{num_epochs}:")
    print(f"  Loss: {epoch_loss:.4f}")
    print(f"  Accuracy: {epoch_acc:.2f}%\n")

print("✓ Training completed!")
````

## 3.8 Visualizing Training Progress

````python
# Plot training progress
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Loss plot
ax1.plot(range(1, num_epochs+1), train_losses, 'b-', marker='o', linewidth=2)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Training Loss Over Time', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

# Accuracy plot
ax2.plot(range(1, num_epochs+1), train_accuracies, 'g-', marker='o', linewidth=2)
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)
ax2.set_title('Training Accuracy Over Time', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_ylim([0, 100])

plt.tight_layout()
plt.show()

print("📊 The model is learning! Loss decreases and accuracy increases.")
````

## 3.9 Evaluating Model Performance

````python
# Evaluate on test set
model.eval()
test_loss = 0
correct = 0
total = 0

with torch.no_grad():
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        test_loss += criterion(output, target).item()
        _, predicted = torch.max(output.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

test_loss = test_loss / len(test_loader)
test_accuracy = 100 * correct / total

print("="*50)
print("FINAL TEST RESULTS")
print("="*50)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.2f}%")
print(f"Correct Predictions: {correct:,} out of {total:,}")
print("="*50)
````

## 3.10 Making Predictions on New Examples

````python
# Get some test samples
model.eval()
test_samples = []
for i in range(10):
    test_samples.append(test_dataset[i])

# Make predictions
fig, axes = plt.subplots(2, 5, figsize=(14, 6))
fig.suptitle('Model Predictions on Test Images', fontsize=14, fontweight='bold')

with torch.no_grad():
    for i, ax in enumerate(axes.flat):
        image, true_label = test_samples[i]
        image_tensor = image.unsqueeze(0).to(device)
        output = model(image_tensor)
        _, predicted = torch.max(output, 1)
        predicted_label = predicted.item()

        # Display
        ax.imshow(image.squeeze(), cmap='gray')

        # Color code: green if correct, red if wrong
        color = 'green' if predicted_label == true_label else 'red'
        ax.set_title(f'True: {true_label} | Pred: {predicted_label}',
                     fontsize=11, color=color, fontweight='bold')
        ax.axis('off')

plt.tight_layout()
plt.show()

print("✓ Green = Correct prediction")
print("✗ Red = Incorrect prediction")
````

## 4.2 From Neural Networks to Large Language Models

The neural network we just built is **specialized** - it only recognizes digits.

**Large Language Models (LLMs)** like GPT-5, Claude, and Gemini are:
- Built on the same fundamental neural network principles
- But **much larger** (billions of parameters vs. our ~100,000)
- Trained on **massive text datasets** (books, websites, articles)
- **General-purpose** - can handle many different tasks

### Key Differences:

| Aspect | Our MLP | Large Language Models |
|--------|---------|----------------------|
| Parameters | ~100,000 | Billions (100M - 1T+) |
| Training Data | 60,000 images | Trillions of words |
| Task | Digit recognition | General text tasks |
| Training Time | Minutes | Weeks/Months |
| Training Cost | Free | Millions of dollars |
| Capabilities | One narrow task | Versatile reasoning |

### But the Core Idea is the Same:
1. **Input**: Data (pixels → text)
2. **Processing**: Multiple layers of neural networks
3. **Learning**: Adjust weights based on errors
4. **Output**: Predictions (digit → next word/answer)


## 4.3 Why LLMs are Transformative for Business

### Traditional Approach:
- Build specific model for each task
- Requires labeled training data
- Needs ML expertise
- Long development time

### LLM Approach:
- One model for many tasks
- Use natural language prompts
- No ML expertise needed
- Immediate deployment

### Business Impact:

**Before LLMs**: To classify customer emails by urgency, you needed:
1. Hire ML engineer
2. Collect labeled email data
3. Train custom model
4. Deploy and maintain
5. **Timeline**: 3-6 months

**With LLMs**: Same task:
1. Write a prompt: "Classify this email as urgent/normal/low priority"
2. **Timeline**: 5 minutes

This democratizes AI - anyone can use it with the right prompts!
